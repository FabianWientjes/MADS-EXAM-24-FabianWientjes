LSTM vs GRU

```python
import matplotlib.pyplot as plt
import numpy as np

# Gegeven data
metrics = [
    "Loss/train",
    "Loss/test",
    "Accuracy",
    "F1scoremicro",
    "F1scoremacro",
    "Precisionmicro",
    "Recallmacro",
]
models = ["LSTM", "GRU"]
loss_train = [0.2711578172736753, 0.15257264222929218]
loss_test = [0.1957554380926821, 0.15956369356976616]
accuracy = [0.928125, 0.9423611111111111]
f1_micro = [0.928125, 0.9423611111111111]
f1_macro = [0.9076189914325637, 0.926056582503554]
precision_micro = [0.928125, 0.9423611111111111]
recall_macro = [0.9121235876715745, 0.9382357429969621]

# Breedte van de staafjes
bar_width = 0.35
index = np.arange(len(metrics))

# Plotting de staafdiagrammen
plt.figure(figsize=(12, 8))

plt.bar(
    index - bar_width / 2,
    [
        loss_train[0],
        loss_test[0],
        accuracy[0],
        f1_micro[0],
        f1_macro[0],
        precision_micro[0],
        recall_macro[0],
    ],
    bar_width,
    label="LSTM",
    color="b",
)
plt.bar(
    index + bar_width / 2,
    [
        loss_train[1],
        loss_test[1],
        accuracy[1],
        f1_micro[1],
        f1_macro[1],
        precision_micro[1],
        recall_macro[1],
    ],
    bar_width,
    label="GRU",
    color="g",
)

# Labels, titels en legenda
plt.xlabel("Metrics")
plt.ylabel("Scores")
plt.title("Vergelijking van metrics tussen LSTM en GRU")
plt.xticks(index, metrics, rotation=45, ha="right")
plt.legend()

# Toon de plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Gegeven data
dropout_values = [0.1, 0.2, 0.3]
hidden_units = [64, 128, 256, 512]
recall_data = np.array(
    [
        [
            0.9451805731874403,
            0.9382357429969621,
            0.9541643268627045,
            0.7929748578634345,
        ],
        [0.9474202386968948, 0.9571853302802544, 0.9466307962996883, 0.5],
        [0.9104507068479077, 0.9125387161618409, 0.9075598938070553, 0.5],
    ]
)

# Plotting the heatmap
plt.figure(figsize=(8, 6))
plt.imshow(recall_data, cmap="YlGnBu", interpolation="nearest")

# Adding ticks and labels
plt.title("Recall Heatmap: Dropout vs. Hidden Units")
plt.xlabel("Hidden Units")
plt.ylabel("Dropout")

plt.xticks(np.arange(len(hidden_units)), hidden_units)
plt.yticks(np.arange(len(dropout_values)), dropout_values)

# Adding the numerical values in the heatmap
for i in range(len(dropout_values)):
    for j in range(len(hidden_units)):
        plt.text(
            j, i, f"{recall_data[i, j]:.2f}", ha="center", va="center", color="black"
        )

# Displaying the plot
plt.colorbar(label="Recall")
plt.tight_layout()
plt.show()
```

Andere metrics met elkaar vergelijken tussen de modellen

```python
# Gegeven data
models = ["2D-convolution", "GRU", "Attention", "GoogleNet"]
loss_test = [0.076, 0.078, 0.067, 0.074]
loss_train = [0.047, 0.055, 0.042, 0.061]
accuracy = [0.981, 0.979, 0.983, 0.981]
f1_score_macro = [0.905, 0.890, 0.913, 0.902]
recall_macro = [0.922, 0.907, 0.922, 0.916]
duration_hours = [1.2, 3.5, 3.8, 3.5]  # Trainingsduur in uren

# Plotting the bar chart
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Loss (test and train)
axes[0, 0].bar(models, loss_test, color="skyblue", label="Loss/test")
axes[0, 0].bar(models, loss_train, color="orange", alpha=0.6, label="Loss/train")
axes[0, 0].set_title("Loss: Test vs Train")
axes[0, 0].set_ylabel("Loss")
axes[0, 0].legend()

# Accuracy
axes[0, 1].bar(models, accuracy, color="green")
axes[0, 1].set_title("Precision (Macro)")
axes[0, 1].set_ylim(0.97, 0.985)  # Setting ylim for better visualization

# F1-score (macro)
axes[1, 0].bar(models, f1_score_macro, color="purple")
axes[1, 0].set_title("F1-score (Macro)")
axes[1, 0].set_ylabel("F1-score")

# Duration subplot
ax_duration = axes[1, 1].bar(models, duration_hours, color="yellow")
axes[1, 1].set_title("Training Duration (hours)")
axes[1, 1].set_ylabel("Hours")

# Adding annotations
for ax in axes.flat:
    for bar in ax.patches:
        ax.annotate(
            f"{bar.get_height():.2f}",
            (bar.get_x() + bar.get_width() / 2, bar.get_height()),
            ha="center",
            va="center",
            xytext=(0, 5),
            textcoords="offset points",
            fontsize=9,
        )

# Adjusting x-labels and layout
for ax in axes.flat:
    ax.set_xticklabels(models, rotation=45, ha="right")
    ax.set_xlabel("Models")

plt.tight_layout()
plt.show()
```

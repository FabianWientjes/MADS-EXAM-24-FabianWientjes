Dropout vs hidden units
Getrained met normale GRU --> Recall macro --> overall 5 epochs

```python
import matplotlib.pyplot as plt
import numpy as np

# Gegeven data
dropout_values = [0.1, 0.2, 0.3]
hidden_units = [128, 256, 512]
recall_data = np.array([
    [0.9382357429969621, 0.9541643268627045, 0.7929748578634345],
    [0.9571853302802544, 0.9466307962996883, 0.5],
    [0.9125387161618409, 0.9075598938070553, 0.5],
])

# Plotting the heatmap
plt.figure(figsize=(8, 6))
plt.imshow(recall_data, cmap='YlGnBu', interpolation='nearest')

# Adding ticks and labels
plt.title('Recall Heatmap: Dropout vs. Hidden Units')
plt.xlabel('Hidden Units')
plt.ylabel('Dropout')

plt.xticks(np.arange(len(hidden_units)), hidden_units)
plt.yticks(np.arange(len(dropout_values)), dropout_values)

# Adding the numerical values in the heatmap
for i in range(len(dropout_values)):
    for j in range(len(hidden_units)):
        plt.text(j, i, f'{recall_data[i, j]:.2f}', ha='center', va='center', color='black')

# Displaying the plot
plt.colorbar(label='Recall')
plt.tight_layout()
plt.show()
```
